```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from keras import Sequential
from keras.layers import Dense , Input
from sklearn.model_selection import train_test_split
```

#Binary Cls

Data

```python
path_1 = pd.read_csv("titanic_data.csv")
```

```python
path_1.head()
```

```python

data = path_1.dropna()
data
```

```python
x_ = data.drop(columns=['survived','alive'])
x_
```

```python
y_ = data[['survived']]
y_
```

```python

x_encoded = pd.get_dummies(x_)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_encoded, y_, test_size=0.2, random_state=42)
```

```python
model = Sequential()
model.add(Input(shape = (X_train.shape[1],))) # Update input shape based on the number of columns after encoding
model.add(Dense(20, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(1, 'sigmoid'))
```

```python
model.compile(optimizer='adam', loss='mse', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, validation_data=(X_test, y_test),epochs=100, batch_size=32)
```
